Membetuk sinyaldigital  dengan t yang berada di range 0-2 dengan interval 0.0001. Dengan t sebagai sumbu waktu dari sinyal.
Membentuk 4 sinyal, dengan rincian sebagai berikut :

![Ketentuan Sinyal](<Ketentuan Sinyal.png>)

```python
import numpy as np
import matplotlib.pyplot as plt

# Membuat array t dari 0 hingga 2 dengan langkah 0.0001
t = np.arange(0, 2, 0.0001)

# Rincian sinyal yang akan dibentuk
y1 = 2 * np.sin(2 * np.pi * 3 * t + 0)
y2 = 1 * np.cos(2 * np.pi * 4 * t + np.pi/4)
y3 = -1 * np.sin(2 * np.pi * 5 * t + np.pi/2)
y4 = 0.5 * np.cos(2 * np.pi * 6 * t + np.pi)

# Buat subplot 2x2
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Plot y1
axs[0, 0].plot(t, y1)
axs[0, 0].set_title("Sinyal y1")
axs[0, 0].grid()

# Plot y2
axs[0, 1].plot(t, y2, color='r')
axs[0, 1].set_title("Sinyal y2")
axs[0, 1].grid()

# Plot y3
axs[1, 0].plot(t, y3, color='g')
axs[1, 0].set_title("Sinyal y3")
axs[1, 0].grid()

# Plot y4
axs[1, 1].plot(t, y4, color='m')
axs[1, 1].set_title("Sinyal y4")
axs[1, 1].grid()

# Penyesuaian layout sinyal
plt.tight_layout()
plt.show()
```

# **Analisis**

### 1. **Besar Amplitudo dan Frekuensi dari masing-masing sinyal yang dibentuk**
  * **Sinyal y1** : Memiliki amplitudo 2 dan frekuensi sebesar 3 Hz
  * **Sinyal y2** : Memiliki amplitudo 1 dan frekuensi sebesar 4 Hz
  * **Sinyal y3** : Memiliki amplitudo -1 dan frekuensi sebesar 5 Hz
  * **Sinyal y4** : Memiliki amplitudo 0.5 dan frekuensi sebesar 6 Hz


### 2. **Bagaimana Fase mempengarhi posisi gelombang**

Fase dalam gelombang sinusoidal menentukan titik awal gelombang pada sumbu waktu (t). Pergeseran fase menyebabkan gelombang bergeser ke kiri jika fase bertambah positif dan bergeser ke kanan jika fase bertambah negatif. 

$$ y_2 = \cos(2\pi \cdot 4 \cdot t + \frac{\pi}{4}) $$


Misalnya pada sinyal y2 tersebut dengan fase $\frac{\pi}{4}$, jika fase diubah mejadi $\frac{\pi}{2}$ maka sinyal akan menjadi lebih maju atau gelombang sinya bergeser ke kiri karena bertambah positif.


### 3. **Dampak atau Pengaruh amplitudo terhadap tampilan gelombang sinyal**

Amplitudo akan memberikan pegaruh berupa seberapa dalam dan tinggi puncak gelombang sinyal yang dihasilkan. Sinyal dengan amplitudo yang lebih tinggi akan terlihat lebih dominan dibadingkan dengan sinyal yang memiliki amplitudo lebih kecil.

Seperti pada **sinyal y1** dengan amplitudo = 2, sinyal yang dihasilkan akan terlihat lebih dalam pada sisi lembah dan lebih tinggi di sisi puncaknya.


### 4. **Perbandingan sinyal dengan pergeseran fase yang berbeda**

Perbandingan akan menggunakan permisalan pada **nomor 2** dengan menggunakan sinyal y2.

$$ y_2 = \cos(2\pi \cdot 4 \cdot t + \frac{\pi}{4}) $$

Misalnya pada sinyal y2 tersebut dengan fase $\frac{\pi}{4}$, jika fase diubah mejadi $\frac{\pi}{2}$, berikut adalah gambar perbandingan ketika dilakukan perubahan fase :

![Perubahan Fase](<Perubahan Fase.png>)

Dari gamber tersebut membuktikan bahwa pergeseran fase akan menentukan posisis awal gelombang sinyal, gelombang sinyal bergeser lebih ke kiri. lembah dan puncak gelombang terlihat lebih awal yang disebabkan oleh fase yang diubah atau ditambah positif.

# **Kombinasi sinyal y5 = y1 + y2**

```python
# Kombinasi sinyal y5 = y1 + y2
y5 = y1 + y2

# Plot y5
plt.figure(figsize=(8, 4))
plt.plot(t, y5, label='y5 = y1 + y2', color='b')
plt.title("Sinyal Kombinasi y5")
plt.xlabel("Waktu (s)")
plt.ylabel("Amplitudo")
plt.grid()
plt.legend()
plt.show()
```

# Analisis
Hasil penjumlahan sinyal menyebabkan gambar atau bentuk sinyal menjadi lebih kompleks, yang mana pola osilasinya berubah-ubah sei setiap waktu. karena terdapat kombinasi amplitudo juga menyebabkan amplitudo yang tidak beraturan dan lebih rumit.
